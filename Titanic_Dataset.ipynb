#Data Cleaning

```python
sns.boxplot(x='Pclass',y='Age',data=df)
```

```python
def age_fill(cols):
  Age=cols[0]
  Pclass=cols[1]

  if pd.isnull(Age):
    if Pclass==1:
      return 37
    elif Pclass==2:
      return 29
    else:
      return 24
  return Age
```

```python
df['Age']=df[['Age','Pclass']].apply(age_fill,axis=1)
```

```python
sns.heatmap(df.isnull())
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

#Build Logistic Regression Model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop('Survived',axis=1),df['Survived'],test_size=0.3,random_state=101)  #this particular code gives all the inputs expect survived
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
log.fit(X_train,y_train)
```

```python
predictions=log.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```

```python
print(classification_report(y_test,predictions))
```
